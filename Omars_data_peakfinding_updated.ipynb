```python
#calibration
de_dic = {218:[50,50],440:[50,20],511:[52,50],1022:[100,100]} #in [,] indicates the left and right hand tolerance for peak finding
kernel = bq.GaussianPeakFilter(500, 50, fwhm_at_0=10)

def pick_peaks(found_peak_array, es, de_dic):
    good_centroid_array = []
    good_es = []

    for i in es:
        rango = de_dic[i]
        diff_exists = False
       
        differ = [x - i for x in found_peak_array]
        differ_neg = [k for k in differ if k < 0]
        differ_pos = [k for k in differ if k >= 0]

        if len(differ_neg)==0:
            differ_neg = [2049]

        if len(differ_pos)==0:
            differ_pos = [2049]

        if (min(np.abs(differ_neg))<rango[0] or min(np.abs(differ_pos))<rango[1]):
            if(min(np.abs(differ_neg))<rango[0] and min(np.abs(differ_pos))>=rango[1]):   
                valid_differ = differ_neg
            if(min(np.abs(differ_neg))>=rango[0] and min(np.abs(differ_pos))<rango[1]):
                valid_differ = differ_pos
            if(min(np.abs(differ_neg))<rango[0] and min(np.abs(differ_pos))<rango[1]):
                valid_differ = differ_neg+differ_pos

            lowest_diff = valid_differ[np.argmin(np.abs(valid_differ))]
            diff_exists = True

        if (diff_exists):
            good_centroid_array.append(lowest_diff+i)
            good_es.append(i)  
    return good_centroid_array, good_es
 
    
#spectrum in calibrate_spec is just a list, so directly from csv

def calibrate_spec(spectrum, kernel, de_dic, es = [218, 440, 511, 1022], min_snr=2, xmin=200, livetime=60.): #es are the peaks of interest for mixed samples
    spec = bq.Spectrum(counts=spectrum, livetime=livetime)
    finder = bq.PeakFinder(spec, kernel)
    finder.find_peaks(min_snr=min_snr, xmin=xmin)
    good_peaks, good_es = pick_peaks(finder.centroids, es, de_dic)
    cal = bq.Calibration.from_points("p[0] + p[1] * x", good_peaks, good_es, params0=[5.0, 0.15])
    spec.apply_calibration(cal)
    bins = spec._bin_edges_kev
    midpt = (bins[1:] + bins[:-1]) / 2
    return midpt, spec
```

class GaussianPeakFilter(PeakFilter):
    """A spectral kernel that is the second derivative of a Gaussian.

    g0(x, u, s) = 1 / (sqrt(2 pi) s) * exp(-(x-u)^2/(2 s^2))
    g1(x, u, s) = -(x - u) / (s^2) * g0
    g2(x, u, s) = (x - u)^2 / (s^4) * g0 - (1/s^2) * g0
                = g0(x,u,s) ((x-u)^2/s^2 - 1) / s^2

    integral(-g2, x=x0..x1)
        = (-g1) for x=x0..x1
        = g1(x0) - g1(x1)

    """

    def kernel(self, x, edges):
        """Generate the kernel for the given x value."""
        fwhm = self.fwhm(x)
        sigma = fwhm / 2.355
        g1_x0 = _gaussian1(edges[:-1], x, sigma)
        g1_x1 = _gaussian1(edges[1:], x, sigma)
        kernel = g1_x0 - g1_x1
        return kernel
