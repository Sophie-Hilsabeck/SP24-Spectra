```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Read the CSV file
url = 'https://raw.githubusercontent.com/Sophie-Hilsabeck/SP24-Spectra/main/spe_rowtry2.csv'
data = pd.read_csv(url)

#Assign numerical indices as time
x = np.arange(1, len(data) + 1)


y = data.iloc[:, 0]  # Assuming counts is in the first column

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')
plt.title('Count vs Channel')
plt.xlabel('Channel')
plt.ylabel('Count')
plt.yscale("log")
plt.grid(True)
plt.show()

#peak finding
peaks, _ = find_peaks(y, height=100)  # Adjust the height as needed

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')
plt.plot(x[peaks], y[peaks], 'x', color='red', markersize=10)
plt.title('Peaks on Count vs Channel')
plt.xlabel('Channel')
plt.yscale("log")
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Print peak positions
print("Peak positions (Channel):", x[peaks].tolist())
print("Peak Count:", y[peaks].tolist())
```
